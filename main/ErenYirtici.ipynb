```python
#Certain Features
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    roc_curve, auc, accuracy_score, f1_score, precision_score, 
    recall_score, matthews_corrcoef, classification_report
)
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import warnings
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

warnings.filterwarnings("ignore", category=FutureWarning)

# Define models with class weights
models = {
    'RandomForest': RandomForestClassifier(class_weight='balanced'),
    'GradientBoosting': GradientBoostingClassifier(),
    'ExtraTrees': ExtraTreesClassifier(class_weight='balanced'),
    'AdaBoost': AdaBoostClassifier(),
    'SVC': SVC(probability=True, class_weight='balanced'),
    'DecisionTree': DecisionTreeClassifier(class_weight='balanced'),
    'NaiveBayes': GaussianNB(),
    'KNN': KNeighborsClassifier(),
    'MLPClassifier': MLPClassifier(max_iter=1000),
    'LDA': LinearDiscriminantAnalysis()
}

# Load and preprocess the dataset
def preprocess_data(filepath):
    data = pd.read_excel(filepath, engine='openpyxl')
    data['CropEstablishment_Merged'] = data['CropEstablishment_CT'] | data['CropEstablishment_CT_line']
    return data

# Evaluate models for specific categories
def evaluate_categories(data, category_columns, target_col, category_name):
    print(f"\n### Starting Evaluation for {category_name} ###")
    for category in category_columns:
        print(f"\n--- Processing Category: {category} ---")
        filtered_data = data[data[category] == 1]

        if filtered_data.empty or filtered_data[target_col].nunique() < 2:
            print(f"Skipping {category} due to insufficient data or class diversity.")
            continue

        X = filtered_data.drop(columns=[target_col])
        y = filtered_data[target_col]

        if y.value_counts().min() < 2:
            print(f"Skipping {category} due to insufficient class samples.")
            continue

        evaluate_models(X, y, category)

# Evaluate models
def evaluate_models(X, y, category):
    smote = SMOTE(random_state=42)
    X_res, y_res = smote.fit_resample(X, y)

    X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

    plt.figure(figsize=(10, 8))

    for name, model in models.items():
        try:
            model.fit(X_train, y_train)
            y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else model.decision_function(X_test)
            fpr, tpr, _ = roc_curve(y_test, y_proba, pos_label=y.unique()[1])
            auc_score = auc(fpr, tpr)

            # Plot ROC curve
            plt.plot(fpr, tpr, label=f"{name} (AUC = {auc_score:.2f})")

            # Print metrics
            y_pred = model.predict(X_test)
            print(f"\nModel: {name}")
            print("AUC:", auc_score)
            print("Accuracy:", accuracy_score(y_test, y_pred))
            print("F1 Score:", f1_score(y_test, y_pred, average='macro', zero_division=0))
            print("Precision:", precision_score(y_test, y_pred, average='macro', zero_division=0))
            print("Recall:", recall_score(y_test, y_pred, average='macro'))
            print("MCC:", matthews_corrcoef(y_test, y_pred))
            print("Classification Report:\n", classification_report(y_test, y_pred))
        except Exception as e:
            print(f"{name} failed: {e}")
            continue

    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve for Category: {category}")
    plt.legend()
    plt.show()

# Main execution
file_path = "C:/Users/Eren/Desktop/yeni/data_4.xlsx"
data = preprocess_data(file_path)

# Define LandType and CropEstablishment categories
landtype_columns = ['LandType_Lowland', 'LandType_MediumLand', 'LandType_Upland']
cropest_columns = ['CropEstablishment_Merged', 'CropEstablishment_ZT']

# Run evaluation
evaluate_categories(data, landtype_columns, target_col='GrainYield', category_name="LandType")
evaluate_categories(data, cropest_columns, target_col='GrainYield', category_name="CropEstablishment")
```

```python
#Feature Selection RFE Method
import pandas as pd
from sklearn.feature_selection import RFE, SelectKBest, mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

def load_and_prepare_data(filepath):
    """Load dataset and encode target variable."""
    data = pd.read_excel(filepath, engine="openpyxl")
    data['GrainYield'] = LabelEncoder().fit_transform(data['GrainYield'])
    return data

def optimized_feature_selection(data, binary_columns, numeric_columns, target_col, n_features=10):
    """Perform optimized feature selection for binary and numeric features."""
    print("\nStarting Optimized Feature Selection...")

    # Split features and target
    X = data.drop(columns=[target_col])
    y = data[target_col]

    selected_features = []

    # Select binary features using RFE
    if binary_columns:
        print("\nProcessing Binary Features...")
        binary_X = X[binary_columns]
        binary_model = RandomForestClassifier(random_state=42, n_estimators=100)
        binary_rfe = RFE(binary_model, n_features_to_select=min(n_features, len(binary_columns)))
        binary_rfe.fit(binary_X, y)
        binary_selected = binary_X.columns[binary_rfe.support_].tolist()
        print(f"Selected Binary Features: {binary_selected}")
        selected_features.extend(binary_selected)

    # Select numeric features using SelectKBest
    if numeric_columns:
        print("\nProcessing Numeric Features...")
        numeric_X = X[numeric_columns]
        selector = SelectKBest(score_func=mutual_info_classif, k=min(n_features, len(numeric_columns)))
        selector.fit(numeric_X, y)
        numeric_selected = numeric_X.columns[selector.get_support()].tolist()
        print(f"Selected Numeric Features: {numeric_selected}")
        selected_features.extend(numeric_selected)

    # Limit to top n_features if more features are selected
    selected_features = selected_features[:n_features]
    print(f"\nFinal Selected Features: {selected_features}")
    return selected_features

# File path
file_path = "C:/Users/Eren/Desktop/yeni/data_4.xlsx"

# Load and prepare data
data = load_and_prepare_data(file_path)

# Define binary and numeric columns
binary_columns = data.iloc[:, 1:85].columns.tolist()  # Assuming first 84 columns after target are binary
numeric_columns = data.iloc[:, 85:].columns.tolist()  # Remaining columns are numeric

# Perform feature selection
top_features = optimized_feature_selection(data, binary_columns, numeric_columns, target_col='GrainYield', n_features=10)

# Display final selected features
print("\nTop Features for Modeling:")
print(top_features)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    roc_curve, auc, accuracy_score, f1_score, precision_score, 
    recall_score, matthews_corrcoef, classification_report
)
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import warnings
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

warnings.filterwarnings("ignore", category=FutureWarning)

# Define models with class weights
models = {
    'RandomForest': RandomForestClassifier(class_weight='balanced'),
    'GradientBoosting': GradientBoostingClassifier(),
    'ExtraTrees': ExtraTreesClassifier(class_weight='balanced'),
    'AdaBoost': AdaBoostClassifier(),
    'SVC': SVC(probability=True, class_weight='balanced'),
    'DecisionTree': DecisionTreeClassifier(class_weight='balanced'),
    'NaiveBayes': GaussianNB(),
    'KNN': KNeighborsClassifier(),
    'MLPClassifier': MLPClassifier(max_iter=1000),
    'LDA': LinearDiscriminantAnalysis()
}

# Load and preprocess the dataset
def preprocess_data(filepath):
    data = pd.read_excel(filepath, engine='openpyxl')
    data['CropEstablishment_Merged'] = data['CropEstablishment_CT'] | data['CropEstablishment_CT_line']
    return data

# Evaluate models for binary columns
def evaluate_binary_columns(data, binary_columns, target_col):
    print("\n### Evaluating Binary Columns ###")
    for col in binary_columns:
        print(f"\n### Processing Binary Column: {col} ###")
        filtered_data = data[data[col] == 1]

        if filtered_data.empty or filtered_data[target_col].nunique() < 2:
            print(f"Skipping {col} due to insufficient data or class diversity.")
            continue

        X = filtered_data.drop(columns=[target_col])
        y = filtered_data[target_col]

        if y.value_counts().min() < 2:
            print(f"Skipping {col} due to insufficient class samples.")
            continue

        evaluate_models(X, y, col)

# Evaluate models for numerical columns
def evaluate_numerical_columns(data, numerical_columns, target_col):
    print("\n### Evaluating Numerical Columns ###")
    for col in numerical_columns:
        print(f"\n### Processing Numerical Column: {col} ###")

        X = data[[col]]
        y = data[target_col]

        if y.value_counts().min() < 2:
            print(f"Skipping {col} due to insufficient class samples.")
            continue

        evaluate_models(X, y, col)

# Evaluate models
def evaluate_models(X, y, category):
    smote = SMOTE(random_state=42)
    X_res, y_res = smote.fit_resample(X, y)

    X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

    plt.figure(figsize=(10, 8))

    best_model = None
    best_score = 0  # Best score tracking for Accuracy

    for name, model in models.items():
        try:
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)  # Accuracy calculation

            # Plot ROC curve
            if hasattr(model, 'predict_proba'):
                y_proba = model.predict_proba(X_test)[:, 1]
                fpr, tpr, _ = roc_curve(y_test, y_proba, pos_label=y.unique()[1])
                auc_score = auc(fpr, tpr)
                plt.plot(fpr, tpr, label=f"{name} (AUC = {auc_score:.2f})")

            # Determine the best model by Accuracy
            if accuracy > best_score:
                best_score = accuracy
                best_model = (name, model)

            # Print performance metrics
            print(f"\nModel: {name}")
            print("Accuracy:", accuracy)
            print("F1 Score:", f1_score(y_test, y_pred, average='macro', zero_division=0))
            print("Precision:", precision_score(y_test, y_pred, average='macro', zero_division=0))
            print("Recall:", recall_score(y_test, y_pred, average='macro'))
            print("MCC:", matthews_corrcoef(y_test, y_pred))
            print("Classification Report:\n", classification_report(y_test, y_pred))
        except Exception as e:
            print(f"{name} failed: {e}")
            continue

    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve for Category: {category}")
    plt.legend()
    plt.show()

    # Draw the confusion matrix of the best model according to Accuracy
    if best_model:
        from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

        best_name, best_model_instance = best_model
        y_pred_best = best_model_instance.predict(X_test)
        cm = confusion_matrix(y_test, y_pred_best, labels=best_model_instance.classes_)

        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_model_instance.classes_)
        disp.plot(cmap="Blues")
        plt.title(f"Confusion Matrix for Best Model: {best_name} (Accuracy: {best_score:.2f})")
        plt.show()



# Main execution
file_path = "C:/Users/Eren/Desktop/yeni/data_4.xlsx"
data = preprocess_data(file_path)

# Define binary and numerical columns
binary_columns = ['District_Lakhisarai', 'LandType_Lowland', 'Variety_HD_2967', 'VarietyClass_LDV', 'VarietyClass_SDV', 'SowingSchedule_T3', 'SowingSchedule_T4', 'SowingSchedule_T5', 'HerbicideName_Total']
numerical_columns = ['Longitude']

target_col = 'GrainYield'

# Run evaluation for binary and numerical columns
evaluate_binary_columns(data, binary_columns, target_col)
evaluate_numerical_columns(data, numerical_columns, target_col)
```
